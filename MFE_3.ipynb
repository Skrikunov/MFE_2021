# Task 1: How is called the general form of SIR system of nonlinear differential equations?
## 1.1) Predator-prey equation analogy:
Classical SIR model is look like this:
 $$
\begin{equation*}
\begin{cases}
\frac{dS}{dt} = - \beta \frac{SI}{N} \\
\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I \\
\frac{dR}{dt} = \gamma I \\
\end{cases}
\end{equation*}
\text{, where}
$$


This linear system is actually the special case of Lotka–Volterra equations: https://en.wikipedia.org/wiki/Lotka%E2%80%93Volterra_equations
The Lotka–Volterra equations, also known as the predator–prey equations, are a pair of first-order nonlinear differential equations, frequently used to describe the dynamics of biological systems in which two species interact, one as a predator and the other as prey. The populations change through time according to the pair of equations:
 $$
\begin{equation*}
\begin{cases}
\frac{dx}{dt} = - \alpha x - \beta xy \\
\frac{dy}{dt} = \delta xy - \gamma y \\
\end{cases}
\end{equation*}
\text{, where}
$$
where

x is the number of prey (for example, rabbits); \
y is the number of some predator (for example, foxes);\
$\frac{dx}{dt}$ and $\frac{dy}{dt}$ represent the instantaneous growth rates of the two populations;\
t represents time;\
α, β, γ, δ are positive real parameters describing the interaction of the two species.\
# 1.2) Kolmogorov equation:
But the predator–prey equations is actually the special case of Kolmogorov equations: https://en.wikipedia.org/wiki/Kolmogorov_equations

 $$
\begin{equation*}
{p_n}^{'}(t) = (n-1)\beta {p_{n-1}}(t)- n\beta {p_{n}}(t) \\
\end{equation*}
$$
This equation is applied to model population growth with birth. Where n is the population index, with reference the initial population, $\beta$  is the birth rate, and finally $p_{n}(t)=\Pr(N(t)=n$, i.e. the probability of achieving a certain population size.

# Age-structured SIR model

The dynamics of the model are given by the following system of differential equations:
$$
\begin{equation*}
\begin{cases}
\frac{dS_i}{dt} = - \beta \frac{S_i}{N} \cdot \sum_{j=1}^{L} M_{ij} I_j \\
\frac{dI_i}{dt} = \beta \frac{S_i}{N} \cdot \sum_{j=1}^{L} M_{ij} I_j - \gamma I_i \\
\frac{dR_i}{dt} = \gamma I_i \\
\end{cases}
\text{, where} \\
\end{equation*}
$$

$N$ - the total population;

$L$ - the number of age groups;

$S_i=S(t_i)$ - the number of susceptible individuals;

$I_i=I(t_i)$ - the number of infectious individuals;

$R_i=R(t_i)$ - the number of removed individuals;

$\textbf{M} \in \mathbb{R}^{L х L}$ - weighted age-contact matrix (describes the rate of contact between each pair of age-brackets);
Values in matrix are proportional to the total number of contacts per time between age-brackets, divided by the product of their population sizes. 

$\beta$ - the transmission rate, measured as the average number of contacts per person per time, multiplied by the probability of transmission between a infected and susceptible person;

$\gamma$ - the removal rate, also given by $\frac{1}{D}$ where $D$ is the length of the period for which a person is infectious

$i$ - iteration number (partiqular time moment).

If we try to transform initial system of equations for $S$ into vector form, we will obtain:

$
\begin{equation*}
\begin{cases}
\frac{dS_1}{dt} = - \beta \frac{S_1}{N} \cdot \sum_{j=1}^{L} M_{1j} I_j \\
\cdot \cdot \cdot \\
\frac{dS_L}{dt} = - \beta \frac{S_L}{N} \cdot \sum_{j=1}^{L} M_{Lj} I_j \\
\end{cases}
\Rightarrow
\frac{d\textbf{S}}{dt} = - \beta \frac{\textbf{S}}{N} \cdot \textbf{MI}
\end{equation*}
$

The same we can do with all others equations ($I,R$). Then we obtain the matrix form of the upper system:

$
\begin{equation*}
\begin{cases}
\frac{d}{dt}\textbf{S} = - \beta \frac{\textbf{S}}{N} \cdot \textbf{MI} \\
\frac{d}{dt}\textbf{I} = \beta \frac{\textbf{S}}{N} \cdot \textbf{MI} - \gamma \textbf{I} \\
\frac{d}{dt}\textbf{R} = \gamma \textbf{I} \\
\end{cases}
\end{equation*}
\text{, where} \\
$

$\textbf{S} \in \mathbb{R}^{L х 1}$ - the vector of susceptible individuals numbers;

$\textbf{I} \in \mathbb{R}^{L х 1}$ - the vector of infectious individuals numbers;

$\textbf{R} \in \mathbb{R}^{L х 1}$ - the vector of removed individuals numbers;

Each of these vectors describes age distribution of particular group of people that are suspected, infectious or removed.

$\beta, \gamma, \textbf{M}$ - the same.

Weighted age-contact matrix $M$ can be calculated from regular $M_0$ matrix by it's multiplication by weighting matrix $P_d$ made from weighting vector $p$:

$$
M = M_0 P_d \\
P_d = diag(\textbf{p}), \text{where } \textbf{p} \in \mathbb{R}^{L х 1}\\
$$

Initially vectors are initialized in the following way:
$
\begin{equation*}
\textbf{S}_0 = N\textbf{p}, \textbf{S}_0(\text{random index})-=1 - \text{to provide at least 1 infecteous} \\
\textbf{I}_0 = N\textbf{p} - \textbf{S}_0\\
\textbf{R}_0 = zeros(L,1)\\
\end{equation*}
$

## The age‑structured SIR model

```python
N = 1e0 # population size

def init_suspected(p,infection_mode,L):
    S0 = N*p # suspected
    # choose random index to be infected
    inf_idx = randint(L)
    if infection_mode=='one_group':
        # to infect some percent of people from the random age group
        S0[inf_idx,:] -= N*to_infect
    elif infection_mode=='all_groups':
        s0 = np.abs(np.random.randn(L,1))
        s0 /= s0.sum()
        # to infect all groups in random distribution
        S0[:,:] -= s0*N*to_infect
    return S0

def SIR_age_structured(beta,gamma,duration,to_infect,infection_mode,scenario):
    # age distribution weighting vector
    ages = age_distributions[scenario]
    print('You selected: ',scenario)
    print('Infection mode: ',infection_mode)
    print('Infected percent: ',to_infect)
    print('beta = ',beta)
    print('gamma = ',gamma)
    print('duration = ',duration)
    print('Initially infected =',to_infect*100,'%')
    
    # the amount of age groups
    L = len(ages)
    
    # normalize weigting vector
    p = np.array(list(ages.values())).reshape(-1,1)/100
    
    # transform weighting vector to weighting matrix
    p_d = np.diag(p[:,0])
    
    # weight initial age-contact matrix
    mmm = MMM0 @ p_d
    
    # initial conditions
    S0 = init_suspected(p,infection_mode,L) # suspected
    I0 = N*p - S0 # infectious
    R0 = np.zeros([L,1]) # removed

    # simulation
    SIR = [[],[],[],[]] # to collect statistics
    # add initial conditions
    SIR[0].append(S0)
    SIR[1].append(I0)
    SIR[2].append(R0)

    # initialize conditions
    S,I,R = S0,I0,R0

    # solve differential equation several times
    for t in range(duration):
        # calculate all equations simultaniously
        S,I,R = S - beta * S * mmm @ I / N, I + beta * S * mmm @ I / N - gamma * I, R + gamma * I

        # collect t-th iteration statistics
        SIR[0].append(S)
        SIR[1].append(I)
        SIR[2].append(R)
        # to be sure if total population is constant
        SIR[3].append(S+I+R)
    return SIR
```

```python
def find_max_xy(x,y):
    val_y=y.max()
    val_x=x[y.argmax()]
    return val_x,val_y

def find_intersection(val,x,y):
    n = int(np.sum(y >= N*val))
    val_x = x[n]
    return val_x

def plot_results(SIR,scenario,titles):
    res=[]
    sum_ = np.array(SIR[3])[:,:,0].sum(axis=1)
    rows = 2
    cols = 3
    plt.subplots(rows,cols,figsize=(20,10))
    counter = 0
    for i in range(rows):
        for j in range(cols):
            plt.subplot(rows, cols, counter+1)
            
            if i == 0:
                total_ = np.array(SIR[j]).sum(axis=1)[:,0] # sum vector along all ages
                title(titles[j],fontsize=fs)
                plot(total_,label=titles[j])
                plot(sum_,label='S+I+R = const')
                legend(loc='upper right')
                if j==0: print('20% intersection = ',find_intersection(0.2,np.arange(len(total_)+1),np.array(total_)))
                if j==1: print('Peak (x,y) = ',find_max_xy(np.arange(len(total_)),np.array(total_)))
                res.append(np.array(total_))
                    
            if i == 1:
                separate_ = np.array(SIR[j])[:,:,0] # consider each row of the vector separately
                plot(separate_)
                legend(age_distributions[scenario].keys())
            counter += 1
            xlabel('Time index',fontsize=fs)
            ylabel('Amount',fontsize=fs)
            grid()
    show()
    return res

titles1 = ['S: susceptible','I: infectious','R: removed']
titles2 = ['H: hospitalized','C: critical care','M: mortality rate']
```

```python
beta = 0.25
gamma = 1/14
duration = 200
to_infect = 0.01
infection_mode = ['one_group','all_groups'][0]
scenario = 'Adams'

SIR = SIR_age_structured(beta,gamma,duration,to_infect,infection_mode,scenario)
res = plot_results(SIR,scenario,titles1)
```

## The age‑structured SIR model + $H,C,M$

$
\begin{equation*}
\begin{cases}
\frac{d}{dt}\textbf{H} = \gamma (\textbf{h} \circ \textbf{I}) \\
\frac{d}{dt}\textbf{C} = \gamma (\textbf{h} \circ \textbf{c} \circ \textbf{I}) \\
\frac{d}{dt}\textbf{M} = \gamma (\textbf{m} \circ \textbf{I}) \\
\end{cases}
\end{equation*}
\text{, where} \\
$

$A \circ B$ - Hadamard product;

$\textbf{h} \in \mathbb{R}^{L х 1}$ - hospitalization rate vector (over age groups distribution);

$\textbf{c} \in \mathbb{R}^{L х 1}$ - ICU (intensive care unit) rate vector among hospitalizations (over age groups distribution);

$\textbf{m} \in \mathbb{R}^{L х 1}$ - mortality rates vector (over age groups distribution);

$\textbf{H} \in \mathbb{R}^{L х 1}$ - total individuals hospitalized;

$\textbf{C} \in \mathbb{R}^{L х 1}$ - total individuals in critical care;

$\textbf{M} \in \mathbb{R}^{L х 1}$ - total individuals deceased respectively;


# Social-distancing policy simulation

On May 29th, Inslee announced a Safe Start: a four phased county-by-county reopening plan. The plan allows counties to gradually relax social-distancing measures based on their assessments of health care system readiness, testing capacity and availability, case and contact investigations, and ability to protect high-risk populations. 

$\textbf{M}_{i+1} = \frac{C_{max}}{\lambda |C|} M_i$, where

$|C|$ - the total number of individuals in critical care;

$C_{max}$ - ICU capacity which (for US average of 34.7 per 100,000 resident);

$\lambda$ - the “strictness” of the social-distancing measures.

This leads to the model, utilizing a rescaling function completely based on the policy that Washington implements where the intensity of the social distancing policy is proportional to the ICU occupancy.

```python
def SIRHCM_age_structured(beta,gamma,duration,to_infect,infection_mode,scenario):
    k=0.01
    h = np.array([[0.1,0.3,1.2,3.2,4.9,10.2,16.6,24.3,27.3]]).T*k
    c = np.array([[5.0,5.0,5.0,5.0,6.3,12.2,27.4,43.2,70.9]]).T*k
    m = np.array([[0.1,0.3,1.2,3.2,4.9,10.2,16.6,24.3,27.3]]).T*k
    lambda_const = 0.1
    C_max = N*(34.7/1e5)

    # age distribution weighting vector
    ages = age_distributions[scenario]
    print('You selected: ',scenario)
    print('Infection mode: ',infection_mode)
    print('Infected percent: ',to_infect)
    print('beta = ',beta)
    print('gamma = ',gamma)
    print('duration = ',duration)
    print('Initially infected =',to_infect*100,'%')
    
    # the amount of age groups
    L = len(ages)
    
    # normalize weigting vector
    p = np.array(list(ages.values())).reshape(-1,1)/100
    
    # transform weighting vector to weighting matrix
    p_d = np.diag(p[:,0])
    
    # weight initial age-contact matrix
    mmm = MMM0 @ p_d
    
    # initial conditions
    S0 = init_suspected(p,infection_mode,L) # suspected
    I0 = N*p - S0 # infectious
    R0 = np.zeros([L,1]) # removed
    H0 = np.zeros([L,1])
    C0 = np.zeros([L,1])
    M0 = np.zeros([L,1])

    # simulation
    SIR = [[],[],[],[]] # to collect statistics
    HCM = [[],[],[],[]]
    C_SUM = [0]
    # add initial conditions
    SIR[0].append(S0)
    SIR[1].append(I0)
    SIR[2].append(R0)
    HCM[0].append(H0)
    HCM[1].append(C0)
    HCM[2].append(M0)
    # initialize conditions
    S,I,R = S0,I0,R0
    H,C,M = H0,C0,M0
    # solve differential equation several times
    I0=np.zeros([L,1])
    for t in range(duration):
        # calculate all equations simultaniously
        S,I,R,H,C,M = S - beta * S * mmm @ I / N, I + beta * S * mmm @ I / N - gamma * I, R + gamma * I, H + gamma*h*(I-I0),C + gamma*h*c*(I-I0), M + gamma*m*(I-I0)
        # in critical care totally
        C_sum = C.sum() 
        C_SUM.append(C_sum)
        scaling_coeff = (C_max)/(lambda_const*C_sum)
        # social-distancing policy simulation
#         mmm = scaling_coeff*mmm
#         I0=I
        
        
        # collect t-th iteration statistics
        SIR[0].append(S)
        SIR[1].append(I)
        SIR[2].append(R)
        SIR[3].append(S+I+R) # to be sure if total population is constant
        HCM[0].append(H)
        HCM[1].append(C)
        HCM[2].append(M)
        HCM[3].append((S+I+R)*0.1)
    return SIR,HCM,C_SUM
```

# Vaccination effect

Age‑targeted vaccination can be described using the following system of equations (modified age-structured SIR model):

$
\begin{equation*}
\begin{cases}
\frac{d}{dt}\textbf{S} = - \beta \frac{\textbf{S}}{N} \cdot \textbf{MI} - {\mu} \\
\frac{d}{dt}\textbf{I} = \beta \frac{\textbf{S}}{N} \cdot \textbf{MI} - \gamma \textbf{I} \\
\frac{d}{dt}\textbf{R} = \gamma \textbf{I} + {\mu}\\
\end{cases}
\end{equation*}
\text{, where} \\
$

$\textbf{S} \in \mathbb{R}^{L х 1}, \textbf{I} \in \mathbb{R}^{L х 1}, \textbf{R} \in \mathbb{R}^{L х 1}, \beta, \gamma, \textbf{M}$ - the same;

$\mu$ - vaccination rate vector ( the number of individuals in each age-bracket vaccinated at each day);

We can model $\mu$ using different vaccination reules that are described by weighting vector $\omega$:

$$
\mu = T \frac{{\omega} \circ \textbf{S}}{|\omega \circ \textbf{S}|}
$$

$T$ -  the total number of vaccines administered at each day $(T = \frac{N}{720})$ - for USA;

## The age‑structured SIR model + vaccination

```python
w_c = np.array([[1,1,1,1,1,1,1,1,1]]).T
w_s = np.array([[1,1,1,1,2,4,8,16,16]]).T 
pd.DataFrame(np.array([w_c,w_s])[:,:,0].T,columns=['w_c','w_s'],index=ages_example.keys())
display(pd.DataFrame(np.array([w_c,w_s])[:,:,0].T,columns=['w_c','w_s'],index=ages_example.keys()).style.set_caption('Vaccination priority vectors'))
```

```python
def SIRV_age_structured(beta,gamma,duration,to_infect,infection_mode,scenario,w):
    # age distribution weighting vector
    ages = age_distributions[scenario]
    print('You selected: ',scenario)
    print('Infection mode: ',infection_mode)
    print('Infected percent: ',to_infect)
    print('beta = ',beta)
    print('gamma = ',gamma)
    print('duration = ',duration)
    print('Initially infected =',to_infect*100,'%')
    
    # the amount of age groups
    L = len(ages)
    
    # normalize weigting vector
    p = np.array(list(ages.values())).reshape(-1,1)/100
    
    # transform weighting vector to weighting matrix
    p_d = np.diag(p[:,0])
    
    # weight initial age-contact matrix
    mmm = MMM0 @ p_d
    
    # initial conditions
    S0 = init_suspected(p,infection_mode,L) # suspected
    I0 = N*p - S0 # infectious
    R0 = np.zeros([L,1]) # removed

    # simulation
    SIR = [[],[],[],[]] # to collect statistics
    # add initial conditions
    SIR[0].append(S0)
    SIR[1].append(I0)
    SIR[2].append(R0)

    # initialize conditions
    S,I,R = S0,I0,R0
    T = N/720
    mu = np.zeros([L,1])
    # solve differential equation several times
    for t in range(duration):
        # calculate all equations simultaniously
        S,I,R,mu = S - (beta * S * mmm @ I / N) - mu, I + (beta * S * mmm @ I / N) - gamma * I, R + (gamma * I) + mu, T * ((w * S) / (w.T @ S))
        
#         S[(S < 0)] = 0
#         I[(I < 0)] = 0
#         R[(R < 0)] = 0

        # collect t-th iteration statistics
        SIR[0].append(S)
        SIR[1].append(I)
        SIR[2].append(R)
        # to be sure if total population is constant
        SIR[3].append(S+I+R)
    return SIR
```

```python
beta = 0.25
gamma = 1/14
duration = 200
to_infect = 0.01
infection_mode = ['one_group','all_groups'][1]
print(list(age_distributions.keys()))
scenario = list(age_distributions.keys())[3]
w_c = np.array([[1,1,1,1,1,1,1,1,1]]).T
w_s = np.array([[1,1,1,1,2,4,8,16,16]]).T 
w_c = w_c/w_c.sum()
w_s = w_s/w_s.sum()

SIRV = SIRV_age_structured(beta,gamma,duration,to_infect,infection_mode,list(age_distributions.keys())[4],w_c)
res1 = plot_results(SIRV,scenario,titles1)
SIRV = SIRV_age_structured(beta,gamma,duration,to_infect,infection_mode,list(age_distributions.keys())[4],w_s)
res2 = plot_results(SIRV,scenario,titles1)
```
